## Is there a relationship between popularity and danceability?

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme('paper')
sns.regplot(x="dnce", y="pop", data=allTopSongs, fit_reg=True)

plt.xlabel('Danceability')
plt.ylabel('Popularity')
 
# title of plot
pvalues = []
plt.title('Popularity vs Danceability')
```

## Does having a higher bpm lead to having a higher energy score?

```python
sns.set_theme('paper')
sns.regplot(x="bpm", y="nrgy", data=allTopSongs, fit_reg=True)

plt.xlabel('BPM')
plt.ylabel('Energy')
 
# title of plot
plt.title('Plot 2')
plt.show()
#correlation table of all numeric data https://www.geeksforgeeks.org/python-pandas-dataframe-corr/?ref=lbp
allTopSongs.corr(method="kendall", min_periods=1, numeric_only=True)
```

If BPM were causally implicated in Energy score I wouldn't expect such vertical variation for any particular BPM.  If almost all BPMs are have a range of energy scores from about 40-90 and there is mild correlation between the two variables, that leads me to concluded that BPM is not causally producing Energy score, but there is a relationship between the two.  If there is a causal i.e. functional relationship between BPM and Energy score I would expect it to be multivariate - including more than one variable to predict energy score with greater accuracy.  For example, while the correlation between bpm and nrgy is .1023, the correlation between spch and nrgy is .1515 which suggests it is more important a factor in determining energy score.  Still I would posit a causal relationship for either.
